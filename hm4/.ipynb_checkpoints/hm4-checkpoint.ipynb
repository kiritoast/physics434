```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

```

```python
t = np.linspace(0,np.pi*100,int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=int(1e5))*np.sqrt(N)+N
```

Problem 1: subsets of data

```python
#partA
plt.plot(t[0:1000], bkgd[0:1000], 'o')
```

```python
#partB
def Generalize_plot(start_point, number_point):
    finish = start_point+number_point
    plt.plot(t[start_point:finish], bkgd[start_point:finish], 'o')
Generalize_plot(500,2000)
```

```python
#partC
plt.plot(t[0:int(1e5):100], bkgd[0:int(1e5):100], 'o')
```

Problem 2: density

```python
#2a
fig, ax = plt.subplots(1, 1)
h=ax.hist2d(t,bkgd,bins=100)
fig.colorbar(h[3], ax=ax, label='Count')
plt.xlabel('time')
plt.ylabel('background data')

plt.show()
```

#2b
The graph is still a scatterplot, with time for x-axis and background data for y-axix. The color works as a 3d histogram. The lighter the color means the higher the bin, which means there are more data point on that position. Opposite color holds true as wells. This graph shows there are many data points in the middle of the graph.

Problem 3: folding

```python
#3a
t_folded = t%(2*(np.pi))
plt.plot(t_folded, bkgd, 'o')
```

```python
#3b
fig, ax = plt.subplots(1, 1)
h=ax.hist2d(t_folded,bkgd,bins=100)
fig.colorbar(h[3], ax=ax, label='Count')
plt.xlabel('time')
plt.ylabel('background data')

plt.show()
```
