```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import math
from scipy.stats import poisson
from scipy.stats import rayleigh
from scipy.stats import norm
```

A) Show how the probability distribution of the background changes as you integrate (sum) for more days.


```python
cosmic = 3
gamma = 5
```

```python
x = np.arange(0,10)
plt.subplot(1, 3, 1)

day1 = stats.poisson.pmf(x,cosmic)
plt.bar(x, day1, fill=True)
plt.title("day 1")

plt.subplot(1, 3, 2)
day2= np.convolve(day1, day1) 
plt.bar(range(len(day2)), day2, fill=True)
plt.title("day 2")

plt.subplot(1, 3, 3)
day3= np.convolve(day2, day2) 
plt.bar(range(len(day3)), day3, fill=True)
plt.title("day 3")

#as sum for more days, the mean slowly increased. It is also more towards the a gaussian dist.
```
