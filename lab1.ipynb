```python
# set hypothetical measurement to 3.002
n = 3.002

# what is the possibility that a data fall into the measure that less than 3.002
# integral from 3 to 3.002 under the orange curve, which is the rayleigh distribution
result=stats.rayleigh.cdf(n,loc = 3., scale = 0.01)

print("the Possibility for n = 3.002 is" ,result)
# sigma = (x-miu)/sd
sigma = (n - 3)/0.01
print("the sigma for n = 3.002 is" ,sigma)
```
