```python
#q2
d = stats.rayleigh.rvs(loc = 3.0, scale = .01, size = 100000)

fig, ax = plt.subplots(1, 1)
plt.tick_params(labelsize = 24)
x = np.linspace(3,3.05,1000)
ax.plot(x,stats.rayleigh.pdf(x,loc = 3., scale = 0.01),linewidth = 8,alpha = 0.7)

plt.show()
```

```python
#q3
```

```python
# set hypothetical measurement to 3.002
n = 3.03
avg = np.mean(d)
# what is the possibility that a data fall into the measure that less than 3.002
# integral from 3 to 3.002 under the orange curve, which is the rayleigh distribution
result=stats.rayleigh.cdf(n,loc = 3., scale = 0.01)


print("the Possibility for n = 3.002 is" ,result)
# sigma = (x-miu)/sd
value=stats.rayleigh.ppf(result,loc = 3., scale = 0.01)

sigma = (value - avg)/0.01
print("the sigma for n = 3.002 is" ,sigma)
#sigma will increase a lot if the the hypothetical measurement values is far off. Also if the standard deviation is smaller,
#the sigma will tends to be big. it is affected by the distance to mean.
```

```python
#For the third graph, what is the possibility that we randomly get a number from poisson distribution that greater 
#than 180 but less than 220.
prob1 = poisson.cdf(220, mu)
prob2 = poisson.cdf(180, mu)
print(prob1 - prob2)
#it would be the difference of Possibility that the number get less than 220 minus the Possibility that the number we get less
#than 180.
```
